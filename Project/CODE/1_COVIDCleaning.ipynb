### This notebook is used to perform **Data Cleaning** on COVID dataset

```python
# Importing the required packages
import csv
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates as mdates
%matplotlib inline
import datetime
from datetime import datetime as dt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
## Process covid data from US Govt website
df = pd.read_csv ("covid_confirmed_usafacts.csv")
#df1 = pd.read_csv ("2020_covid.csv")
print("read complete")
```

```python
df.head()
```

```python
## Remove the word county from county name to match the Accident dataset
df['CountyName'] = df['CountyName'].str.replace(' County', '')
df.head()
```

```python
##Remove spaces to match the accident file

df.CountyName = df.CountyName.str.rstrip()
df.CountyName = df.CountyName.str.lstrip()
df.State = df.State.str.rstrip()
df.State = df.State.str.rstrip()
```

```python
#check for the word county in data
df[df['CountyName'].str.contains ("County")]
```

```python
##Drop unwanted fields
df=df.drop(columns=['StateFIPS','countyFIPS'])
df.head()
```

```python
##convert the columnar data for every date into row wise data
idx = ["CountyName","State"]             # identifier variable
new_df=pd.melt(df, id_vars=idx, var_name="Date",value_name="Count").sort_values(idx).reset_index(drop=True)
```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df.dtypes
new_df.head()
```
